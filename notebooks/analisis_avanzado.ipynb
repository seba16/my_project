```python
DATA_PATH = "../data/retail_sales_dataset.csv"
import sys
sys.path.append('../src')
import pandas_helper
import pandas as pd
print(f"Intentando leer archivo desde: {DATA_PATH}")
df = pandas_helper.read_csv_data(DATA_PATH)

df = pandas_helper.clean_data(df)
```

```python
df.head()
```

```python
# porcentaje de la compra de cada customer_id según la categoria de producto
df['percentage_of_purchase'] = df.groupby('customer_id')['product_category'].transform('count') / df['product_category'].count()
df.head()
```

```python
# clasificacion de compras por customer_id. Si la venta es mayor a 2000, se clasifica como "high_value_purchase", si es menor a 1000, se clasifica como "low_value_purchase", else "medium_value_purchase"
df['purchase_classification'] = pd.cut(df['total_amount'], bins=[0, 500, 1500, float('inf')], labels=['low_value_purchase', 'medium_value_purchase', 'high_value_purchase'])
df
```

```python
# agrupación multiple gender, product_category, purchase_classification
df_grouped = df.groupby(['gender', 'product_category', 'purchase_classification']).agg({'total_amount': 'sum'})
df_grouped
```

```python
# Aplicar funciones de agregación: Utiliza funciones como sum, mean, count, min, max, std, y var para obtener estadísticas descriptivas de cada grupo
# la mediana de las ventas catalogadas como "high_value_purchase"
df_high_value_purchase = df[df['purchase_classification'] == 'high_value_purchase'] # se filtran las ventas catalogadas como "high_value_purchase"
float(df_high_value_purchase['total_amount'].mean()) # se calcula la mediana con el metodo mean
```

```python
# el maximo de las ventas catalogadas como "high_value_purchase"
df_high_value_purchase = df[df['purchase_classification'] == 'high_value_purchase'] # se filtran las ventas "high_value_purchase"
int(df_high_value_purchase['total_amount'].max()) # se obtiene el maximo de las ventas "high_value_purchase"
```

```python
# el minimo de las ventas catalogadas como "high_value_purchase"
df_high_value_purchase = df[df['purchase_classification'] == 'high_value_purchase'] # se filtran las ventas "high_value_purchase"
int(df_high_value_purchase['total_amount'].min()) # se obtiene el minimo de las ventas "high_value_purchase"
```

```python
# cantidad de ventas catalogadas como "medium_value_purchase"
df_medium_value_purchase = df[df['purchase_classification'] == 'medium_value_purchase'] # se filtran las ventas "medium_value_purchase"
int(df_medium_value_purchase['total_amount'].count()) # se obtiene la cantidad de ventas "medium_value_purchase"
```

```python
# suma de las ventas catalogadas como "low_value_purchase"
df_low_value_purchase = df[df['purchase_classification'] == 'low_value_purchase'] # se filtran las ventas "low_value_purchase"
int(df_low_value_purchase['total_amount'].sum()) # se obtiene la suma de las ventas "low_value_purchase"
```

```python
# desviacion estandar de las ventas catalogadas como "medium_value_purchase"
df_medium_value_purchase = df[df['purchase_classification'] == 'medium_value_purchase'] # se filtran las ventas "medium_value_purchase"
float(df_medium_value_purchase['total_amount'].std()) # se obtiene la desviación estandar
```

```python
# varianza de las ventas catalogadas como "low_value_purchase"
df_low_value_purchase = df[df['purchase_classification'] == 'low_value_purchase']# se filtran las ventas "low_value_purchase"
float(df_low_value_purchase['total_amount'].var()) # se obtiene la varianza
```

```python
# utilizar función apply para calcular grupo etario
# Si la edad es menor a 30, se clasifica como "joven", si es mayor a 30 y menor a 60, se clasifica como "adulto", else "adulto_mayor"
df['age_group'] = df['age'].apply(
    lambda x: 'joven' if x < 30 else ('adulto' if x < 60 else 'adulto_mayor')
) 
df
```

```python
# Calcula la desviación de cada venta respecto a la media de su grupo.
df['deviation'] = df.groupby('age_group')['total_amount'].transform(lambda x: x - x.mean())
df
```
